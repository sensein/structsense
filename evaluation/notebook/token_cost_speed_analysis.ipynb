## Reproschema Token Usage Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv("reproschema/reproschema_token_usage.csv")
```

```python
# Optional: Clean whitespace from 'Model' names
df['Model'] = df['Model'].str.strip()

# Set style and palette
sns.set(style="whitegrid")
sns.set_palette("Set2")

# Plot 1: Cost per Model (Violin Plot)
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Model', y='Cost ($)', inner='box')
plt.title("Cost Distribution by Model")
plt.tight_layout()
plt.savefig("reproschema/reproschema_cost_violin.svg")
plt.show()

# Plot 2: Input and Output Tokens per Model (Violin Plot)
plt.figure(figsize=(10, 6))
df_long = df.melt(id_vars=['Model'], value_vars=['Input Tokens', 'Output Tokens'],
                  var_name='Token Type', value_name='Count')
sns.violinplot(data=df_long, x='Model', y='Count', hue='Token Type', split=True)
plt.title("Token Usage by Model")
plt.legend(title='Token Type', title_fontsize='13', fontsize='11')  # Increased legend font sizes
plt.tight_layout()
plt.savefig("reproschema/reproschema_token_usage_violin.svg")
plt.show()

# Plot 3: Speed per Model (Violin Plot)
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Model', y='Speed (tps)', inner='box')
plt.title("Speed (tokens/sec) Distribution by Model")
plt.tight_layout()
plt.savefig("reproschema/reproschema_speed_violin.svg")
plt.show()

# Plot 4: Scatter plot of Speed vs Cost
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Speed (tps)', y='Cost ($)', hue='Model', style='Model', s=100)
plt.title("Speed vs Cost by Model")
plt.legend(title='Model', title_fontsize='13', fontsize='11')  # Increased legend font sizes
plt.tight_layout()
plt.savefig("reproschema/reproschema_speed_vs_cost.svg")
plt.show()
```

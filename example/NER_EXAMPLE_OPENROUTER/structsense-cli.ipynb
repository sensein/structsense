- output is always returned in JSON, you can specify the keys/structure (optional)
- we use some pre-defined variables, which will be replaced with the actual value during the run-time.
  - literature -- in extractor agent. 
  - extracted_structured_information - in alignment agent.
  - aligned_structured_information -- in judge agent.
  - judged_structured_information_with_human_feedback -- in human feedback agent.
  - modification_context -- in human feedback agent.
